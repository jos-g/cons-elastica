```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mpl
```

```python
X = np.array([0.002,0.005,0.007,0.010,0.012,0.015,0.017,0.020,0.022,0.025]) # em kilogramas
Y = [0.01,0.06,0.09,0.13,0.15,0.20,0.22,0.27,0.30,0.35] # em m
# Y = (g/k)*m
# g = 9.8 m/s²
# k no SI.
```

```python
plt.figure( figsize=(8,8))
plt.plot(X,Y, 'ko')
plt.xlabel( 'massa (kg)')
plt.ylabel( 'L (m)')

```

```python
mod_linear = np.polyfit(X, Y , 1) 
a = float(mod_linear[0])
b = float(mod_linear[1])
print("coeficiente angular da reta: {:.2f}".format(a))
print("coeficiente linear da reta: {:.2f}".format(b))
```

```python

k = a/9.8 #m .s-2

print("A constante da mola é {:.2f} N/m.".format(k))
```

```python
y_mod = a*X + b
print(y_mod)
```

```python
plt.figure( figsize=(8,8))
plt.plot(X,Y, "ko", label= 'Dados Coletados')
plt.plot(X,y_mod, "r--", label='Reta do Ajuste Linear')
plt.xlabel("massa (kg)")
plt.ylabel("L (m)")
plt.legend()
plt.show()
```

```python
#R²
yresid = Y - y_mod
SQresid = sum(yresid**2)
y_tot = Y- np.mean(Y)
SQtotal = sum(y_tot**2)
R2 = 1 - SQresid/SQtotal
print ('coeficiente de determinação do ajuste é {:.2f}'.format(R2))
```

```python
M = np.array([0.002,0.005,0.007,0.010,0.012,0.015,0.017,0.020,0.022,0.025]) # em kilogramas
t = np.array([5.50,5.97,6.56,7.13,8.19,8.82,7.30,9.30,9.87,10])
#T² = (4pi²/k)*m
# onde T = t/20 
```

```python
plt.figure( figsize=(8,8))
plt.plot(M,t, 'ko')
plt.xlabel( 'massa (kg)')
plt.ylabel( 'T² (s²)')
```

```python
import math
```

```python
T1 = t/10
print(T1)
```

```python
p = math.pi
k = 2*p
```

```python
T = np.array(T1/k)**2
print(T)
```

```python
mod_linear2 = np.polyfit(M, T , 1) 
a2 = float(mod_linear2[0])
b2 = float(mod_linear2[1])
print("coeficiente angular da reta: {:.2f}".format(a2))
print("coeficiente linear da reta: {:.2f}".format(b2))
```

```python

K= (k)**2/a2
print("o valor de K é {:.2f} kg/s².".format(k))
```

```python
y_mod2 = a2*M + b2
print(y_mod2)
```

```python
#diagrama de dispersão
plt.figure( figsize=(8,8))
plt.plot(M,T, "ko", label='Dados Coletados')
plt.plot(M,y_mod2, "r--", label = 'Reta do ajuste linear')
plt.xlabel("m (kg)")
plt.ylabel("(T/2π)² (s²)")
plt.legend()
plt.show()
```

```python
#R²
yresid2 = T - y_mod2
SQresid2 = sum(yresid2**2)
y_tot2 = T- np.mean(T)
SQtotal2 = sum(y_tot2**2)
R_2 = 1 - SQresid2/SQtotal2
print ('coeficiente de determinação do ajuste é {:.2f}'.format(R_2))
```
